```python
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = '/content/Vulnerability_to_Displacement_by_NPA (1).csv'
df = pd.read_csv(file_path)

print("DataFrame Info:")
df.info()

print("\nFirst 5 rows of the DataFrame:")
display(df.head())
```

### Cleaning Redundant Columns, Missing Values, and Data Types

```python
# Drop the 'Pct' columns as their numeric counterparts
df = df.drop(columns=['PctHighSchool', 'PctNonWhite', 'Pct65_', 'PctHH_Poverty', 'PctHome_Ownership'])

# Drop rows with any remaining missing values
df.dropna(inplace=True)

# Convert 'Vulnerable' column from 'Yes'/'No' strings to boolean (True/False)
df['Vulnerable'] = df['Vulnerable'].map({'Yes': True, 'No': False})

print("DataFrame Info after cleaning:")
df.info()

print("\nFirst 5 rows of DataFrame after cleaning:")
display(df.head())
```

### Reordering Columns for Readability

```python
ordered_columns = [
    'OBJECTID', 'ID', 'GlobalID', 'NPA',
    'HighSchool', 'NonWhite', 'F65_', 'Poverty', 'Home_Ownership',
    'Score', 'Vulnerable',
    'ShapeSTArea', 'ShapeSTLength'
]

# reorder the DataFrame columns
df = df[ordered_columns]

print("\nDataFrame head after reordering columns:")
display(df.head())
```
